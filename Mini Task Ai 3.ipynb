```python
import pandas as pd

# Load dataset with correct headers
df = pd.read_csv("iris.csv", header=None)

# Assign column names manually for Iris dataset
df.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]

print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Distribution of Sepal Length
sns.histplot(df["sepal_length"], kde=True, color="blue")
plt.title("Distribution of Sepal Length")
plt.show()

# 2. Average Sepal vs Petal by Species
sns.barplot(x="species", y="sepal_length", data=df, palette="Set2")
plt.title("Average Sepal Length by Species")
plt.show()

# 3. Boxplot of Petal Length by Species
sns.boxplot(x="species", y="petal_length", data=df, palette="muted")
plt.title("Petal Length Distribution by Species")
plt.show()

# 4. Violin plot (Sepal Width vs Species)
sns.violinplot(x="species", y="sepal_width", data=df, palette="coolwarm")
plt.title("Sepal Width by Species")
plt.show()

# 5. Pairplot (All features by Species)
sns.pairplot(df, hue="species", palette="husl")
plt.show()
```
